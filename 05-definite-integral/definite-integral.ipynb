```python
from sympy import * 
x, t = symbols('x t', real=True)
T = symbols('T', real=True, positive=True)
```

```python
# There is no built-in odd function or even function in sympy.
# So we use known odd functions for the demo
# The integral of an odd function over [-T,T] is 0
odd_function = sin(x)
integrate(odd_function,(x,-T,T))
```

```python
# The integral of an even function over [-T,T]
even_function = cos(x)
simplify(integrate(even_function,(x,-T,T)))
```

```python
# Gaussian distribution
# We know that guassian distribution takes the form
mu = symbols('mu', real=True)
sigma = symbols('sigma', real=True, positive=True)
h = exp(-(x-mu)**2/sigma**2/2)
h
```

```python
# The integral of it over in the real domain should be 1
# We call this the normalization
A = integrate(h,(x,-oo,oo))
A
```

```python
# So the normalized distribution is 
h1 = h/A
h1
```

```python
plot(h1.subs(mu,0).subs(sigma,1))
```

```python
# Its integral from -oo to xi
xi = Symbol('xi', real=True)
error_function = simplify(integrate(h1,(x,-oo,xi)))
error_function
```

```python
plot(error_function.subs(mu,0).subs(sigma,1), (xi,-10,10))
```
